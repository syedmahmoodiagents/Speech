```python
import librosa
import numpy as np
import soundfile as sf
```

```python
from pydub import AudioSegment
```

```python
import torch
from transformers import WhisperProcessor, WhisperForConditionalGeneration
```

```python
model_name = "openai/whisper-small"
```

```python
import os, getpass
```

```python
os.environ['HF_TOKEN'] = getpass.getpass('Enter HF token: ')
```

```python
processor = WhisperProcessor.from_pretrained(model_name)
model = WhisperForConditionalGeneration.from_pretrained(model_name)
```

```python
AudioSegment.from_mp3("audio_sample.mp3").export("input.wav", format="wav")
```

```python
audio_path = "input.wav"
audio, sr = librosa.load(audio_path, sr=16000)
```

```python
inputs = processor(audio, sampling_rate=16000, return_tensors="pt")
```

```python
# inputs['input_features'].shape
```

```python
# Generate translation (set task + target language)
forced_decoder_ids = processor.get_decoder_prompt_ids(
    language="fr",  # target language
    task="translate"
)
```

```python
with torch.no_grad():
    predicted_ids = model.generate(
        inputs.input_features,
        forced_decoder_ids=forced_decoder_ids
    )
```

```python
# Decode output
translation = processor.batch_decode(
    predicted_ids,
    skip_special_tokens=True
)
```

```python
print("Translated text:", translation[0])
```
